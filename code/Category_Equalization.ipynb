### Json 파일을 읽어들여 카테고리를 균일하게 분포시키는 코드

```python
# 모듈 로딩
import glob, json, os, shutil
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from pprint import pprint
from pylabel import importer
```

```python
# annotation 폴더 내의 json 파일 불러오기
LABEL_PATH = r'C:/DataScience/T3Q/data/DownsamplingData/TrainingData/folder/Annotation/'
IMAGE_PATH = r'C:/DataScience/T3Q/data/DownsamplingData/TrainingData/folder/images/'
LABEL_SAVE_PATH = r'C:/DataScience/T3Q/data/DownsamplingData/TrainingData/folder/equalized_label/'
IMAGE_SAVE_PATH = r'C:/DataScience/T3Q/data/DownsamplingData/TrainingData/folder/equalized_image/'
json_list = os.listdir(LABEL_PATH)
img_list = os.listdir(IMAGE_PATH)
```

```python
cat_1 = 0  # Animals(Dolls)                                    delete
cat_2 = 0  # Person
cat_3 = 0  # Garbage bag & sacks
cat_4 = 0  # Construction signs & Parking prohibited board
cat_5 = 0  # Traffic cone
cat_6 = 0  # Box                                               delete
cat_7 = 0  # Stones on road                                    delete
cat_8 = 0  # Pothole on road
cat_9 = 0  # Filled pothole
cat_10 = 0  # Manhole
cat_list = ["Animals(Dolls)", "Person", "Garbage bag & sacks", "Construction signs & Parking prohibited board",
            "Traffic cone",
            "Box", "Stones on road", "Pothole on road", "Filled pothole", "Manhole"]
```
